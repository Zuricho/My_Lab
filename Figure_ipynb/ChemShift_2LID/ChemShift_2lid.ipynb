```python
# Load 4 kinds of chemiNl shift data, all secondary chemiNl shift

NShift2_sim1 = LoadPred_N("2lid.md.pdb_pred_sim1.tab")
NShift2_sim2 = LoadPred_N("2lid.md.pdb_pred_sim2.tab")
NShift2_sim3 = LoadPred_N("2lid.md.pdb_pred_sim3.tab")
NShift2_sim4 = LoadPred_N("2lid.md.pdb_pred_sim4.tab")
NShift2_sim5 = LoadPred_N("2lid.md.pdb_pred_sim5.tab")

AvgShift = [(NShift2_sim1[i]+NShift2_sim2[i]+NShift2_sim3[i]+NShift2_sim4[i]+NShift2_sim5[i])/5 if NShift2_sim1[i] != None else None for i in range(len(NShift2_sim1))]
AvgShift = [AvgShift[i]-NShift_RC_sim1[i] if AvgShift[i] != None and NShift_RC_sim1[i] != None else None for i in range(len(NShift_RC_sim1))]

NShift2_CMAP_sim1 = LoadPred_N("2lid.md.pdb_pred_CMAP_sim1.tab")
NShift2_CMAP_sim2 = LoadPred_N("2lid.md.pdb_pred_CMAP_sim2.tab")
NShift2_CMAP_sim3 = LoadPred_N("2lid.md.pdb_pred_CMAP_sim3.tab")
NShift2_CMAP_sim4 = LoadPred_N("2lid.md.pdb_pred_CMAP_sim4.tab")
NShift2_CMAP_sim5 = LoadPred_N("2lid.md.pdb_pred_CMAP_sim5.tab")

AvgShift_CMAP = [(NShift2_CMAP_sim1[i]+NShift2_CMAP_sim2[i]+NShift2_CMAP_sim3[i]+NShift2_CMAP_sim4[i]+NShift2_CMAP_sim5[i])/5 if NShift2_CMAP_sim1[i] != None else None for i in range(len(NShift2_CMAP_sim1))]
AvgShift_CMAP = [AvgShift_CMAP[i]-NShift_RC_sim1[i] if AvgShift_CMAP[i] != None and NShift_RC_sim1[i] != None else None for i in range(len(NShift_RC_sim1))]


NShift2_D_sim1 = LoadPred_N("2lid.md.pdb_pred_D_sim1.tab")
NShift2_D_sim2 = LoadPred_N("2lid.md.pdb_pred_D_sim2.tab")
NShift2_D_sim3 = LoadPred_N("2lid.md.pdb_pred_D_sim3.tab")
NShift2_D_sim4 = LoadPred_N("2lid.md.pdb_pred_D_sim4.tab")
NShift2_D_sim5 = LoadPred_N("2lid.md.pdb_pred_D_sim5.tab")

AvgShift_D = [(NShift2_D_sim1[i]+NShift2_D_sim2[i]+NShift2_D_sim3[i]+NShift2_D_sim4[i]+NShift2_D_sim5[i])/5 if NShift2_D_sim1[i] != None else None for i in range(len(NShift2_D_sim1))]
AvgShift_D = [AvgShift_D[i]-NShift_RC_sim1[i] if AvgShift_D[i] != None and NShift_RC_sim1[i] != None else None for i in range(len(NShift_RC_sim1))]


NShift2_D_CMAP_sim1 = LoadPred_N("2lid.md.pdb_pred_D_CMAP_sim1.tab")
NShift2_D_CMAP_sim2 = LoadPred_N("2lid.md.pdb_pred_D_CMAP_sim2.tab")
NShift2_D_CMAP_sim3 = LoadPred_N("2lid.md.pdb_pred_D_CMAP_sim3.tab")
NShift2_D_CMAP_sim4 = LoadPred_N("2lid.md.pdb_pred_D_CMAP_sim4.tab")
NShift2_D_CMAP_sim5 = LoadPred_N("2lid.md.pdb_pred_D_CMAP_sim5.tab")

AvgShift_D_CMAP = [(NShift2_D_CMAP_sim1[i]+NShift2_D_CMAP_sim2[i]+NShift2_D_CMAP_sim3[i]+NShift2_D_CMAP_sim4[i]+NShift2_D_CMAP_sim5[i])/5 if NShift2_D_CMAP_sim1[i] != None else None for i in range(len(NShift2_D_CMAP_sim1))]
AvgShift_D_CMAP = [AvgShift_D_CMAP[i]-NShift_RC_sim1[i] if AvgShift_D_CMAP[i] != None and NShift_RC_sim1[i] != None else None for i in range(len(NShift_RC_sim1))]
```

```python
# Plotting!

plt.subplots(figsize=(10,4))

plt.plot(AvgShift,'.-',label='ffptm TIP4Pew',color="tab:blue")
plt.plot(AvgShift_CMAP,'.-',label='ff03CMAP TIP4Pew',color="tab:orange")
plt.plot(AvgShift_D,'.-',label='ffptm TIP4PD',color="tab:green")
plt.plot(AvgShift_D_CMAP,'.-',label='ff03CMAP TIP4PD',color="tab:red")

plt.plot(ExpData,'.-',label='Experiment',color="black")



plt.title('N Chemical shift of 2LID',loc='left')

plt.xlim(0, len(AvgShift)-1)
plt.xticks(np.arange(0,len(res_list)),res_list,rotation='60')

# Tag modified residues
plt.axvline(x=11, ymin=0, ymax=1,color="grey",linestyle="--")

plt.legend()
plt.savefig("ChemShift_2lid.png",dpi=100)
plt.show()
```
