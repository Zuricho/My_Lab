```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

def loadfile(fileName):
    content = np.loadtxt(fileName,skiprows=1)
    return content

```

```python
dihAngle_EW = loadfile("phipsi_EW.dat")
dihAngle_EW_CMAP = loadfile("phipsi_EW_CMAP.dat")

dihAngle_D = loadfile("phipsi_D.dat")
dihAngle_D_CMAP = loadfile("phipsi_D_CMAP.dat")



plt.subplots(figsize=(4, 2))

plt.xticks()
plt.yticks([])

sns.kdeplot(dihAngle_D[:,1],shade=True,cut=0,color="tab:green")

plt.xlim(-180,180)

# sns.distplot(plotData)

plt.show()
```

```python
distance_EW = loadfile("distance_EW.dat")
distance_EW_CMAP = loadfile("distance_EW_CMAP.dat")

distance_D = loadfile("distance_D.dat")
distance_D_CMAP = loadfile("distance_D_CMAP.dat")



plt.subplots(figsize=(4, 2))

plt.xticks()
plt.yticks([])

sns.kdeplot(distance_D[:,1],shade=True,cut=0,color="tab:green")

plt.xlim(0,)

# sns.distplot(plotData)

plt.show()
```
