```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

def loadfile(fileName):
    content = np.loadtxt(fileName,skiprows=1)
    return content

```

```python
dihAngle_EW = loadfile("phipsi_EW.dat")
dihAngle_EW_CMAP = loadfile("phipsi_EW_CMAP.dat")

dihAngle_D = loadfile("phipsi_D.dat")
dihAngle_D_CMAP = loadfile("phipsi_D_CMAP.dat")



plt.subplots(figsize=(10,8))
plt.suptitle("9 Phosphoserine",fontsize='x-large',fontweight='bold')

plt.xticks()
plt.yticks([])

plt.subplot(221)
sns.kdeplot(dihAngle_EW[:,1],shade=True,cut=0,color="tab:blue")
sns.kdeplot(dihAngle_EW_CMAP[:,1],shade=True,cut=0,color="tab:orange")
plt.title("TIP4PEW phi",fontweight='bold')
plt.xticks()
plt.yticks([])
plt.xlim(-180,180)


plt.subplot(222)
sns.kdeplot(dihAngle_EW[:,2],shade=True,cut=0,color="tab:blue",label="forcefield PTM")
sns.kdeplot(dihAngle_EW_CMAP[:,2],shade=True,cut=0,color="tab:orange",label="ff03CMAP")
plt.title("TIP4PEW psi",fontweight='bold')
plt.xticks()
plt.yticks([])
plt.xlim(-180,180)
plt.legend()


plt.subplot(223)
sns.kdeplot(dihAngle_D[:,1],shade=True,cut=0,color="tab:green")
sns.kdeplot(dihAngle_D_CMAP[:,1],shade=True,cut=0,color="tab:red")
plt.title("TIP4PD phi",fontweight='bold')
plt.xticks()
plt.yticks([])
plt.xlim(-180,180)


plt.subplot(224)
sns.kdeplot(dihAngle_D[:,2],shade=True,cut=0,color="tab:green",label="forcefield PTM")
sns.kdeplot(dihAngle_D_CMAP[:,2],shade=True,cut=0,color="tab:red",label="ff03CMAP")
plt.title("TIP4PD psi",fontweight='bold')
plt.xticks()
plt.yticks([])
plt.xlim(-180,180)
plt.legend()



# sns.distplot(plotData)
plt.savefig("9sep.png",dpi=100)
plt.show()
```
