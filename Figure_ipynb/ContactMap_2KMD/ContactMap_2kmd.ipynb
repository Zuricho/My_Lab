```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
NatContEW = np.loadtxt("native.map_2kmd_EW.dat")
NonContEW = np.loadtxt("nonnative.map_2kmd_EW.dat")

NatContEWCMAP = np.loadtxt("native.map_2kmd_EW_CMAP.dat")
NonContEWCMAP = np.loadtxt("nonnative.map_2kmd_EW_CMAP.dat")

NatContD = np.loadtxt("native.map_2kmd_D.dat")
NonContD = np.loadtxt("nonnative.map_2kmd_D.dat")

NatContDCMAP = np.loadtxt("native.map_2kmd_D_CMAP.dat")
NonContDCMAP = np.loadtxt("nonnative.map_2kmd_D_CMAP.dat")
```

```python
ContMapEW = NatContEW[:,2] + NonContEW[:,2]
ContMapEW = ContMapEW.reshape((111,111))

ContMapEWCMAP = NatContEWCMAP[:,2] + NonContEWCMAP[:,2]
ContMapEWCMAP = ContMapEWCMAP.reshape((111,111))

ContMapD = NatContD[:,2] + NonContD[:,2]
ContMapD = ContMapD.reshape((111,111))

ContMapDCMAP = NatContDCMAP[:,2] + NonContDCMAP[:,2]
ContMapDCMAP = ContMapDCMAP.reshape((111,111))


NonContMapEW = NonContEW[:,2]
NonContMapEW = NonContMapEW.reshape((111,111))

NonContMapEWCMAP = NonContEWCMAP[:,2]
NonContMapEWCMAP = NonContMapEWCMAP.reshape((111,111))

NonContMapD = NonContD[:,2]
NonContMapD = NonContMapD.reshape((111,111))

NonContMapDCMAP = NonContDCMAP[:,2]
NonContMapDCMAP = NonContMapDCMAP.reshape((111,111))
```

```python
plt.subplots(figsize=(8,8))

plt.subplot(221)
plt.imshow(ContMapEW,cmap='jet')

plt.subplot(222)
plt.imshow(ContMapEWCMAP,cmap='jet')

plt.subplot(223)
plt.imshow(ContMapD,cmap='jet')

plt.subplot(224)
plt.imshow(ContMapDCMAP,cmap='jet')

plt.tight_layout(pad=0.1)

plt.savefig("Native_2kmd.png",dpi=300)
plt.show()
```

```python
plt.subplots(figsize=(8,8))

plt.subplot(221)
plt.imshow(NonContMapEW,cmap='jet')

plt.subplot(222)
plt.imshow(NonContMapEWCMAP,cmap='jet')

plt.subplot(223)
plt.imshow(NonContMapD,cmap='jet')

plt.subplot(224)
plt.imshow(NonContMapDCMAP,cmap='jet')

plt.tight_layout(pad=0.1)

plt.savefig("NonNative_2kmd.png",dpi=300)
plt.show()
```
