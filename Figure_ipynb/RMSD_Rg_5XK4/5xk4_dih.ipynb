```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


def loadfile_Rg(fileName):
    content = np.loadtxt(fileName,skiprows=1)
    return content[:,1]

def loadfile_RMSD(fileName):
    content = np.loadtxt(fileName,skiprows=1)
    return content[:,1]

```

```python

Rg_1 = loadfile_Rg("Rg_5xk4_sim1.dat")
Rg_2 = loadfile_Rg("Rg_5xk4_sim2.dat")
Rg_3 = loadfile_Rg("Rg_5xk4_sim3.dat")
Rg_4 = loadfile_Rg("Rg_5xk4_sim4.dat")
Rg_5 = loadfile_Rg("Rg_5xk4_sim5.dat")


RMSD_1 = loadfile_RMSD("RMSD_5xk4_sim1.dat")
RMSD_2 = loadfile_RMSD("RMSD_5xk4_sim2.dat")
RMSD_3 = loadfile_RMSD("RMSD_5xk4_sim3.dat")
RMSD_4 = loadfile_RMSD("RMSD_5xk4_sim4.dat")
RMSD_5 = loadfile_RMSD("RMSD_5xk4_sim5.dat")


Rg_data_EW = np.append(Rg_1,[Rg_2,Rg_3,Rg_4,Rg_5])
RMSD_data_EW = np.append(RMSD_1,[RMSD_2,RMSD_3,RMSD_4,RMSD_5])
```

```python
Rg_1 = loadfile_Rg("Rg_5xk4_CMAP_sim1.dat")
Rg_2 = loadfile_Rg("Rg_5xk4_CMAP_sim2.dat")
Rg_3 = loadfile_Rg("Rg_5xk4_CMAP_sim3.dat")
Rg_4 = loadfile_Rg("Rg_5xk4_CMAP_sim4.dat")
Rg_5 = loadfile_Rg("Rg_5xk4_CMAP_sim5.dat")


RMSD_1 = loadfile_RMSD("RMSD_5xk4_CMAP_sim1.dat")
RMSD_2 = loadfile_RMSD("RMSD_5xk4_CMAP_sim2.dat")
RMSD_3 = loadfile_RMSD("RMSD_5xk4_CMAP_sim3.dat")
RMSD_4 = loadfile_RMSD("RMSD_5xk4_CMAP_sim4.dat")
RMSD_5 = loadfile_RMSD("RMSD_5xk4_CMAP_sim5.dat")


Rg_data_EW_CMAP = np.append(Rg_1,[Rg_2,Rg_3,Rg_4,Rg_5])
RMSD_data_EW_CMAP = np.append(RMSD_1,[RMSD_2,RMSD_3,RMSD_4,RMSD_5])
```

```python
Rg_1 = loadfile_Rg("Rg_5xk4_D_sim1.dat")
Rg_2 = loadfile_Rg("Rg_5xk4_D_sim2.dat")
Rg_3 = loadfile_Rg("Rg_5xk4_D_sim3.dat")
Rg_4 = loadfile_Rg("Rg_5xk4_D_sim4.dat")
Rg_5 = loadfile_Rg("Rg_5xk4_D_sim5.dat")


RMSD_1 = loadfile_RMSD("RMSD_5xk4_D_sim1.dat")
RMSD_2 = loadfile_RMSD("RMSD_5xk4_D_sim2.dat")
RMSD_3 = loadfile_RMSD("RMSD_5xk4_D_sim3.dat")
RMSD_4 = loadfile_RMSD("RMSD_5xk4_D_sim4.dat")
RMSD_5 = loadfile_RMSD("RMSD_5xk4_D_sim5.dat")


Rg_data_D = np.append(Rg_1,[Rg_2,Rg_3,Rg_4,Rg_5])
RMSD_data_D = np.append(RMSD_1,[RMSD_2,RMSD_3,RMSD_4,RMSD_5])
```

```python
Rg_1 = loadfile_Rg("Rg_5xk4_D_CMAP_sim1.dat")
Rg_2 = loadfile_Rg("Rg_5xk4_D_CMAP_sim2.dat")
Rg_3 = loadfile_Rg("Rg_5xk4_D_CMAP_sim3.dat")
Rg_4 = loadfile_Rg("Rg_5xk4_D_CMAP_sim4.dat")
Rg_5 = loadfile_Rg("Rg_5xk4_D_CMAP_sim5.dat")


RMSD_1 = loadfile_RMSD("RMSD_5xk4_D_CMAP_sim1.dat")
RMSD_2 = loadfile_RMSD("RMSD_5xk4_D_CMAP_sim2.dat")
RMSD_3 = loadfile_RMSD("RMSD_5xk4_D_CMAP_sim3.dat")
RMSD_4 = loadfile_RMSD("RMSD_5xk4_D_CMAP_sim4.dat")
RMSD_5 = loadfile_RMSD("RMSD_5xk4_D_CMAP_sim5.dat")


Rg_data_D_CMAP = np.append(Rg_1,[Rg_2,Rg_3,Rg_4,Rg_5])
RMSD_data_D_CMAP = np.append(RMSD_1,[RMSD_2,RMSD_3,RMSD_4,RMSD_5])
```

```python
plt.subplots(figsize=(12, 12))

plt.subplot(221)
im = sns.kdeplot(Rg_data_EW,RMSD_data_EW,cut=0,n_levels=50,shade=True,cmap="CMRmap_r",vmax=5,cbar=True)
plt.title('Forcefield PTM TIP4PEW',fontweight='bold')

plt.subplot(222)
sns.kdeplot(Rg_data_EW_CMAP,RMSD_data_EW_CMAP,cut=0,n_levels=50,shade=True,cmap="CMRmap_r",vmax=5,cbar=True)
plt.title('ff03CMAP TIP4PEW',fontweight='bold')


plt.subplot(223)
sns.kdeplot(Rg_data_D,RMSD_data_D,cut=0,n_levels=50,shade=True,cmap="CMRmap_r",vmax=5,cbar=True)
plt.title('Forcefield PTM TIP4PD',fontweight='bold')

plt.subplot(224)
sns.kdeplot(Rg_data_D_CMAP,RMSD_data_D_CMAP,cut=0,n_levels=50,shade=True,cmap="CMRmap_r",vmax=5,cbar=True)
plt.title('ff03CMAP TIP4PD',fontweight='bold')

plt.savefig("RMSD_Rg_TIP4PEW.png",dpi=300)

plt.show()
```
