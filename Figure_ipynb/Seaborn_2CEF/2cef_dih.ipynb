```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


def loadfile_Rg(fileName):
    content = np.loadtxt(fileName,skiprows=1)
    return content[:,1]

def loadfile_RMSD(fileName):
    content = np.loadtxt(fileName,skiprows=1)
    return content[:,1]

```

```python

Rg_1 = loadfile_Rg("Rg_2cef_sim1.dat")
Rg_2 = loadfile_Rg("Rg_2cef_sim2.dat")
Rg_3 = loadfile_Rg("Rg_2cef_sim3.dat")
Rg_4 = loadfile_Rg("Rg_2cef_sim4.dat")
Rg_5 = loadfile_Rg("Rg_2cef_sim5.dat")


RMSD_1 = loadfile_RMSD("RMSD_2cef_sim1.dat")
RMSD_2 = loadfile_RMSD("RMSD_2cef_sim2.dat")
RMSD_3 = loadfile_RMSD("RMSD_2cef_sim3.dat")
RMSD_4 = loadfile_RMSD("RMSD_2cef_sim4.dat")
RMSD_5 = loadfile_RMSD("RMSD_2cef_sim5.dat")


Rg_data = np.append(Rg_1,[Rg_2,Rg_3,Rg_4,Rg_5])
RMSD_data = np.append(RMSD_1,[RMSD_2,RMSD_3,RMSD_4,RMSD_5])



plt.subplots(figsize=(4, 4))

sns.kdeplot(Rg_data,RMSD_data,cut=0,n_levels=100,shade=True,cmap="Reds")

plt.title('forcefield PTM TIP4PEW')
plt.xlabel('Rg(Å)')
plt.ylabel('RMSD(Å)')

plt.savefig("RMSD_Rg.png",dpi=100)

plt.show()
```
