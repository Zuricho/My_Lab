```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
NatContEW = np.loadtxt("native.map_2cef_EW.dat")
NonContEW = np.loadtxt("nonnative.map_2cef_EW.dat")

NatContEWCMAP = np.loadtxt("native.map_2cef_EW_CMAP.dat")
NonContEWCMAP = np.loadtxt("nonnative.map_2cef_EW_CMAP.dat")

NatContD = np.loadtxt("native.map_2cef_D.dat")
NonContD = np.loadtxt("nonnative.map_2cef_D.dat")

NatContDCMAP = np.loadtxt("native.map_2cef_D_CMAP.dat")
NonContDCMAP = np.loadtxt("nonnative.map_2cef_D_CMAP.dat")
```

```python
ContMapEW = NatContEW[:,2] + NonContEW[:,2]
ContMapEW = ContMapEW.reshape((19,19))

ContMapEWCMAP = NatContEWCMAP[:,2] + NonContEWCMAP[:,2]
ContMapEWCMAP = ContMapEWCMAP.reshape((19,19))

ContMapD = NatContD[:,2] + NonContD[:,2]
ContMapD = ContMapD.reshape((19,19))

ContMapDCMAP = NatContDCMAP[:,2] + NonContDCMAP[:,2]
ContMapDCMAP = ContMapDCMAP.reshape((19,19))


NonContMapEW = NonContEW[:,2]
NonContMapEW = NonContMapEW.reshape((19,19))

NonContMapEWCMAP = NonContEWCMAP[:,2]
NonContMapEWCMAP = NonContMapEWCMAP.reshape((19,19))

NonContMapD = NonContD[:,2]
NonContMapD = NonContMapD.reshape((19,19))

NonContMapDCMAP = NonContDCMAP[:,2]
NonContMapDCMAP = NonContMapDCMAP.reshape((19,19))
```

```python
plt.subplots(figsize=(8,8))

plt.subplot(221)
plt.imshow(ContMapEW,cmap='jet')
plt.xticks(np.arange(0,19))
plt.yticks(np.arange(0,19))

plt.subplot(222)
plt.imshow(ContMapEWCMAP,cmap='jet')
plt.xticks(np.arange(0,19))
plt.yticks(np.arange(0,19))

plt.subplot(223)
plt.imshow(ContMapD,cmap='jet')
plt.xticks(np.arange(0,19))
plt.yticks(np.arange(0,19))

plt.subplot(224)
plt.imshow(ContMapDCMAP,cmap='jet')
plt.xticks(np.arange(0,19))
plt.yticks(np.arange(0,19))

plt.tight_layout(pad=0.1)

plt.savefig("Native_2cef.png",dpi=300)
plt.show()
```

```python
plt.subplots(figsize=(8,8))

plt.subplot(221)
plt.imshow(NonContMapEW,cmap='jet')
plt.xticks(np.arange(0,19))
plt.yticks(np.arange(0,19))

plt.subplot(222)
plt.imshow(NonContMapEWCMAP,cmap='jet')
plt.xticks(np.arange(0,19))
plt.yticks(np.arange(0,19))

plt.subplot(223)
plt.imshow(NonContMapD,cmap='jet')
plt.xticks(np.arange(0,19))
plt.yticks(np.arange(0,19))

plt.subplot(224)
plt.imshow(NonContMapDCMAP,cmap='jet')
plt.xticks(np.arange(0,19))
plt.yticks(np.arange(0,19))

plt.tight_layout(pad=0.1)

plt.savefig("NonNative_2cef.png",dpi=300)
plt.show()
```
