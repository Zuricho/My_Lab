```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


def loadfile_Rg(fileName):
    content = np.loadtxt(fileName,skiprows=1)
    return content[:,1]

def loadfile_RMSD(fileName):
    content = np.loadtxt(fileName,skiprows=1)
    return content[:,1]

```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats


def loadfile_Rg(fileName):
    content = np.loadtxt(fileName,skiprows=1)
    return content[:,1]

def loadfile_RMSD(fileName):
    content = np.loadtxt(fileName,skiprows=1)
    return content[:,1]

def get_extent(Rg_data_list,RMSD_data_list):
    if len(Rg_data_list) != len(RMSD_data_list):
        raise BaseException("Error: Rg and RMSD data list length not equal")
    else:
        xmin = []
        xmax = []
        ymin = []
        ymax = []
        for i in range(len(Rg_data_list)):
            xmin.append(Rg_data_list[i].min())
            xmax.append(Rg_data_list[i].max())
            ymin.append(RMSD_data_list[i].min())
            ymax.append(RMSD_data_list[i].max())
        return min(xmin), max(xmax), min(ymin), max(ymax)

def get_gaussian_kde(Rg_data,RMSD_data,xmin,xmax,ymin,ymax):
    m1, m2 = Rg_data,RMSD_data
    X, Y = np.mgrid[xmin:xmax:100j, ymin:ymax:100j]
    positions = np.vstack([X.ravel(), Y.ravel()])
    values = np.vstack([m1, m2])
    kernel = stats.gaussian_kde(values)
    Z = np.reshape(kernel(positions), X.shape).T
    return Z

```

```python
fig, axs = plt.subplots(2,2,figsize=(10,8))

contour_levels = np.linspace(0,0.4,300)
colorbar_levels = np.arange(0,0.4+0.01,0.1)


xmin, xmax, ymin, ymax = get_extent([Rg_data_EW,Rg_data_D,Rg_data_EW_CMAP,Rg_data_D_CMAP],[RMSD_data_EW,RMSD_data_D,RMSD_data_EW_CMAP,RMSD_data_D_CMAP])



ax = axs[0,0]
Z = get_gaussian_kde(Rg_data_EW,RMSD_data_EW,xmin, xmax, ymin, ymax)
cont = ax.contourf(Z,
            levels=contour_levels,
            cmap=plt.cm.RdBu_r,
            extent=[xmin, xmax, ymin, ymax],
            extend='both')
ax.set_xlim([xmin, xmax])
ax.set_ylim([ymin, ymax])
ax.text(x=xmax-(xmax-xmin)/20,y=ymin+(ymax-ymin)/10,s="ff03/TIP4PEw",fontsize=16,color="white",ha="right",va="center")
ax.tick_params(labelsize=14)
ax.set_ylabel("RMSD ()",fontsize=14)


ax = axs[0,1]
Z = get_gaussian_kde(Rg_data_EW_CMAP,RMSD_data_EW_CMAP,xmin, xmax, ymin, ymax)
cont = ax.contourf(Z,
            levels=contour_levels,
            cmap=plt.cm.RdBu_r,
            extent=[xmin, xmax, ymin, ymax],
            extend='both')
ax.set_xlim([xmin, xmax])
ax.set_ylim([ymin, ymax])
ax.text(x=xmax-(xmax-xmin)/20,y=ymin+(ymax-ymin)/10,s="ff03CMAP/TIP4PEw",fontsize=16,color="white",ha="right",va="center")
ax.tick_params(labelsize=14)


ax = axs[1,0]
Z = get_gaussian_kde(Rg_data_D,RMSD_data_D,xmin, xmax, ymin, ymax)
cont = ax.contourf(Z,
            levels=contour_levels,
            cmap=plt.cm.RdBu_r,
            extent=[xmin, xmax, ymin, ymax],
            extend='both')
ax.set_xlim([xmin, xmax])
ax.set_ylim([ymin, ymax])
ax.text(x=xmax-(xmax-xmin)/20,y=ymin+(ymax-ymin)/10,s="ff03/TIP4PD",fontsize=16,color="white",ha="right",va="center")
ax.tick_params(labelsize=14)
ax.set_xlabel("Rg ()",fontsize=14)
ax.set_ylabel("RMSD ()",fontsize=14)


ax = axs[1,1]
Z = get_gaussian_kde(Rg_data_D_CMAP,RMSD_data_D_CMAP,xmin, xmax, ymin, ymax)
cont = ax.contourf(Z,
            levels=contour_levels,
            cmap=plt.cm.RdBu_r,
            extent=[xmin, xmax, ymin, ymax],
            extend='both')
ax.set_xlim([xmin, xmax])
ax.set_ylim([ymin, ymax])
ax.text(x=xmax-(xmax-xmin)/20,y=ymin+(ymax-ymin)/10,s="ff03CMAP/TIP4PD",fontsize=16,color="white",ha="right",va="center")
ax.tick_params(labelsize=14)
ax.set_xlabel("Rg ()",fontsize=14)


cbar = plt.colorbar(cont,ax=axs,shrink=0.8,ticks=colorbar_levels)
cbar.ax.tick_params(labelsize=12)
cbar.set_label(label='Density',fontsize=14)



plt.savefig("RMSD_Rg_2cef_kde.png",dpi=200)
plt.show()
```
