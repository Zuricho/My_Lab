```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
hbondEW = np.loadtxt("hbond_2lid_EW.dat",skiprows=1)
hbondEWCMAP = np.loadtxt("hbond_2lid_EW_CMAP.dat",skiprows=1)
hbondD = np.loadtxt("hbond_2lid_D.dat",skiprows=1)
hbondDCMAP = np.loadtxt("hbond_2lid_D_CMAP.dat",skiprows=1)
```

```python
plt.subplots(figsize=(6,4))

Bins = np.arange(-0.5,max(hbondEW[:,1])+1.5,1)
HistdataEW = plt.hist(hbondEW[:,1],bins=Bins,visible=False)
plt.plot(np.arange(max(hbondEW[:,1])+1),HistdataEW[0],'.-',color='tab:blue',label='ffPTM TIP4PEW')

Bins = np.arange(-0.5,max(hbondEWCMAP[:,1])+1.5,1)
HistdataEWCMAP = plt.hist(hbondEWCMAP[:,1],bins=Bins,visible=False)
plt.plot(np.arange(max(hbondEWCMAP[:,1])+1),HistdataEWCMAP[0],'.-',color='tab:orange',label='ff03CMAP TIP4PEW')

Bins = np.arange(-0.5,max(hbondD[:,1])+1.5,1)
HistdataD = plt.hist(hbondD[:,1],bins=Bins,visible=False)
plt.plot(np.arange(max(hbondD[:,1])+1),HistdataD[0],'.-',color='tab:green',label='ffPTM TIP4PD')

Bins = np.arange(-0.5,max(hbondDCMAP[:,1])+1.5,1)
HistdataDCMAP = plt.hist(hbondDCMAP[:,1],bins=Bins,visible=False)
plt.plot(np.arange(max(hbondDCMAP[:,1])+1),HistdataDCMAP[0],'.-',color='tab:red',label='ff03CMAP TIP4PD')

plt.xlabel("Number of H-bonds")
plt.ylabel("Frequency")

plt.show()

plt.close()
```

```python
plt.subplots(figsize=(6,4))

Bins = np.arange(-0.5,max(hbondEW[:,2]+hbondEW[:,3])+1.5,1)
HistdataEW_M = plt.hist(hbondEW[:,2]+hbondEW[:,3],bins=Bins,visible=False)
plt.plot(np.arange(max(hbondEW[:,2]+hbondEW[:,3])+1),HistdataEW_M[0],'.-',color='tab:blue',label='ffPTM TIP4PEW')

Bins = np.arange(-0.5,max(hbondEWCMAP[:,2]+hbondEWCMAP[:,3])+1.5,1)
HistdataEWCMAP_M = plt.hist(hbondEWCMAP[:,2]+hbondEWCMAP[:,3],bins=Bins,visible=False)
plt.plot(np.arange(max(hbondEWCMAP[:,2]+hbondEWCMAP[:,3])+1),HistdataEWCMAP_M[0],'.-',color='tab:orange',label='ff03CMAP TIP4PEW')

Bins = np.arange(-0.5,max(hbondD[:,2]+hbondD[:,3])+1.5,1)
HistdataD_M = plt.hist(hbondD[:,2]+hbondD[:,3],bins=Bins,visible=False)
plt.plot(np.arange(max(hbondD[:,2]+hbondD[:,3])+1),HistdataD_M[0],'.-',color='tab:green',label='ffPTM TIP4PD')

Bins = np.arange(-0.5,max(hbondDCMAP[:,2]+hbondDCMAP[:,3])+1.5,1)
HistdataDCMAP_M = plt.hist(hbondDCMAP[:,2]+hbondDCMAP[:,3],bins=Bins,visible=False)
plt.plot(np.arange(max(hbondDCMAP[:,2]+hbondDCMAP[:,3])+1),HistdataDCMAP_M[0],'.-',color='tab:red',label='ff03CMAP TIP4PD')

plt.xlabel("Number of H-bonds")
plt.ylabel("Frequency")

plt.show()

plt.close()
```

```python
plt.subplots(figsize=(8,3))

plt.subplot(121)
plt.plot(np.arange(max(hbondEW[:,1])+1),HistdataEW[0]/25000,'.-',color='tab:blue',label='ff03 TIP4PEW')
plt.plot(np.arange(max(hbondEWCMAP[:,1])+1),HistdataEWCMAP[0]/25000,'.-',color='tab:orange',label='ff03CMAP TIP4PEW')
plt.plot(np.arange(max(hbondD[:,1])+1),HistdataD[0]/25000,'.-',color='tab:green',label='ff03 TIP4PD')
plt.plot(np.arange(max(hbondDCMAP[:,1])+1),HistdataDCMAP[0]/25000,'.-',color='tab:red',label='ff03CMAP TIP4PD')

plt.xlim=(0,)
plt.ylim=(0,)

plt.xlabel("Number of H-bonds")
plt.ylabel("Probability")

plt.subplot(122)

plt.plot(np.arange(max(hbondEW[:,2]+hbondEW[:,3])+1),HistdataEW_M[0]/25000,'.-',color='tab:blue',label='ff03 TIP4PEW')
plt.plot(np.arange(max(hbondEWCMAP[:,2]+hbondEWCMAP[:,3])+1),HistdataEWCMAP_M[0]/25000,'.-',color='tab:orange',label='ff03CMAP TIP4PEW')
plt.plot(np.arange(max(hbondD[:,2]+hbondD[:,3])+1),HistdataD_M[0]/25000,'.-',color='tab:green',label='ff03 TIP4PD')
plt.plot(np.arange(max(hbondDCMAP[:,2]+hbondDCMAP[:,3])+1),HistdataDCMAP_M[0]/25000,'.-',color='tab:red',label='ff03CMAP TIP4PD')

plt.xlim=(0,)
plt.ylim=(0,)

plt.xlabel("Number of H-bonds with modified site")

plt.legend()
plt.tight_layout(pad=0.1)

plt.savefig('hbond_2lid.png',dpi=200)

plt.show()
```
