```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib


def LoadDssp(fileName):
    dssp = np.loadtxt(fileName,skiprows=1)
    return [dssp[i,3]+dssp[i,4]+dssp[i,5] for i in range(len(dssp))]


def LoadD2d(fileName):
    f = open(fileName)
    r = f.readlines()
    content = []
    for i in range(28,len(r)):
        content.append(r[i].split())
    return content
```

```python
dsspEW = LoadDssp("dssp_sum_EW.dat")
dsspEWC = LoadDssp("dssp_sum_EW_CMAP.dat")
dsspD = LoadDssp("dssp_sum_D.dat")
dsspDC = LoadDssp("dssp_sum_D_CMAP.dat")
```

```python
ExpData = [0.023,None,0.157,0.177,0.108,0.060,0.031,0.019,0.030,0.073,0.137,None,0.063,0.029,0.030]
```

```python
plt.subplots(figsize=(6,3))

plt.title("Fraction Helical 2CEF")

plt.plot(ExpData,'.',label="Exp",color='black')
plt.plot(dsspEW,label="ffptm TIP4PEW")
plt.plot(dsspEWC,label="ff03CMAP TIP4PEW")
plt.plot(dsspD,label="ffptm TIP4PD")
plt.plot(dsspDC,label="ff03CMAP TIP4PD")

plt.ylim(0,0.5)

plt.savefig('DSSP_2cef.png',dpi=100)

plt.legend()
```

```python
dssp_EW = np.loadtxt("dssp_sum_EW.dat",skiprows=1)
dssp_EWC = np.loadtxt("dssp_sum_EW_CMAP.dat",skiprows=1)
dssp_D = np.loadtxt("dssp_sum_D.dat",skiprows=1)
dssp_DC = np.loadtxt("dssp_sum_D_CMAP.dat",skiprows=1)
```

```python
plt.subplots(figsize=(7,5))

plt.title("Secondary Structure 2CEF ffPTM TIP4PEW")

dataX = np.arange(len(dssp_EW))
bottomX = np.zeros(len(dssp_EW))
finalX = np.ones(len(dssp_EW))

plt.bar(dssp_EW[:,0],dssp_EW[:,1],bottom=bottomX,width=0.7,color='tab:blue',label='Para')
bottomX += dssp_EW[:,1]

plt.bar(dssp_EW[:,0],dssp_EW[:,2],bottom=bottomX,width=0.7,color='tab:orange',label='Anti')
bottomX += dssp_EW[:,2]

plt.bar(dssp_EW[:,0],dssp_EW[:,3],bottom=bottomX,width=0.7,color='tab:green',label='3-10')
bottomX += dssp_EW[:,3]

plt.bar(dssp_EW[:,0],dssp_EW[:,4],bottom=bottomX,width=0.7,color='tab:red',label='Alpha')
bottomX += dssp_EW[:,4]

plt.bar(dssp_EW[:,0],dssp_EW[:,5],bottom=bottomX,width=0.7,color='tab:purple',label='Pi')
bottomX += dssp_EW[:,5]

plt.bar(dssp_EW[:,0],dssp_EW[:,6],bottom=bottomX,width=0.7,color='tab:brown',label='Turn')
bottomX += dssp_EW[:,6]

plt.bar(dssp_EW[:,0],dssp_EW[:,7],bottom=bottomX,width=0.7,color='tab:pink',label='Bend')
bottomX += dssp_EW[:,7]

plt.bar(dssp_EW[:,0],finalX-bottomX,bottom=bottomX,width=0.7,color='tab:gray',label='Disorder')

plt.ylim(0,1)
plt.xlim(0.5,len(dssp_EW)+0.5)
plt.ylabel("Secondary structure fraction")
plt.xticks(dssp_EW[:,0])
plt.xlabel("Residues")

plt.legend(loc='lower left',bbox_to_anchor=(1,0))
plt.tight_layout()

plt.savefig("SecStruct_2cef.png",dpi=300)

plt.show()
```
