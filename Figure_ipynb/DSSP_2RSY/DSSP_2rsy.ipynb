```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
```

```python
dsspEW = np.loadtxt("dssp_sum_2rsy_EW.dat",skiprows=1)
dsspEWC = np.loadtxt("dssp_sum_2rsy_EW_CMAP.dat",skiprows=1)
dsspD = np.loadtxt("dssp_sum_2rsy_D.dat",skiprows=1)
dsspDC = np.loadtxt("dssp_sum_2rsy_D_CMAP.dat",skiprows=1)
```

```python
plt.subplots(figsize=(24,6))

colorX = [None,'purple','royalblue','cyan','lime','yellow','orange','tomato']
labelX = [None,'Para','Anti','3-10','Alpha','Pi','Turn','Bend']

plt.tight_layout()

# ffPTM TIP4PEW
plt.subplot(221)

dataX = np.arange(len(dsspEW))
bottomX = np.zeros(len(dsspEW))
finalX = np.ones(len(dsspEW))

for i in range(1,8):
    plt.bar(dsspEW[:,0],dsspEW[:,i],bottom=bottomX,width=0.8,color=colorX[i],label=labelX[i])
    bottomX += dsspEW[:,i]

plt.bar(dsspEW[:,0],finalX-bottomX,bottom=bottomX,width=0.8,color='black',label='Disorder')

plt.ylim(0,1)
plt.xlim(0.5,len(dsspEW)+0.5)
plt.ylabel("Secondary structure fraction")
plt.xticks([])


# ff03CMAP TIP4PEW
plt.subplot(222)

dataX = np.arange(len(dsspEWC))
bottomX = np.zeros(len(dsspEWC))
finalX = np.ones(len(dsspEWC))

for i in range(1,8):
    plt.bar(dsspEWC[:,0],dsspEWC[:,i],bottom=bottomX,width=0.8,color=colorX[i],label=labelX[i])
    bottomX += dsspEWC[:,i]

plt.bar(dsspEWC[:,0],finalX-bottomX,bottom=bottomX,width=0.8,color='black',label='Disorder')

plt.ylim(0,1)
plt.xlim(0.5,len(dsspEWC)+0.5)
plt.xticks([])
plt.yticks([])


# ffPTM TIP4PD
plt.subplot(223)

dataX = np.arange(len(dsspD))
bottomX = np.zeros(len(dsspD))
finalX = np.ones(len(dsspD))

for i in range(1,8):
    plt.bar(dsspD[:,0],dsspD[:,i],bottom=bottomX,width=0.8,color=colorX[i],label=labelX[i])
    bottomX += dsspD[:,i]

plt.bar(dsspD[:,0],finalX-bottomX,bottom=bottomX,width=0.8,color='black',label='Disorder')

plt.ylim(0,1)
plt.xlim(0.5,len(dsspD)+0.5)
plt.ylabel("Secondary structure fraction")
plt.xticks(np.arange(1,137,3))
plt.xlabel("Residues")


# ff03PTM TIP4PD
plt.subplot(224)

dataX = np.arange(len(dsspDC))
bottomX = np.zeros(len(dsspDC))
finalX = np.ones(len(dsspDC))

for i in range(1,8):
    plt.bar(dsspDC[:,0],dsspDC[:,i],bottom=bottomX,width=0.8,color=colorX[i],label=labelX[i])
    bottomX += dsspDC[:,i]

plt.bar(dsspDC[:,0],finalX-bottomX,bottom=bottomX,width=0.8,color='black',label='Disorder')

plt.ylim(0,1)
plt.xlim(0.5,len(dsspDC)+0.5)
plt.xticks(np.arange(1,137,10))
plt.yticks([])
plt.xlabel("Residues")


plt.legend(loc='lower left',bbox_to_anchor=(1,0))
plt.tight_layout(pad=0.1)

plt.savefig("SecStruct_2rsy.png",dpi=300)

plt.show()
```

```python
dsspTotalEW = np.loadtxt("dssp_2rsy_EW.dat",skiprows=1)

time = dsspTotalEW[:,0]
content = dsspTotalEW[:,1:]

colorTList = ['black','purple','royalblue','cyan','lime','yellow','orange','tomato']
colorT = matplotlib.colors.ListedColormap(colorTList)
labelT = ['Disorder','Para','Anti','3-10','Alpha','Pi','Turn','Bend']

plt.subplots(figsize=(20,6))

plt.subplot(151)
plt.imshow(content.T[:,:5000],aspect='auto',cmap=colorT)
plt.xticks([0,1000,2000,3000,4000])
plt.yticks(np.arange(0,137,10))

plt.subplot(152)
plt.imshow(content.T[:,5000:10000],aspect='auto',cmap=colorT)
plt.xticks([])
plt.yticks([])

plt.subplot(153)
plt.imshow(content.T[:,10000:15000],aspect='auto',cmap=colorT)
plt.xticks([])
plt.yticks([])

plt.subplot(154)
plt.imshow(content.T[:,15000:20000],aspect='auto',cmap=colorT)
plt.xticks([])
plt.yticks([])

plt.subplot(155)
plt.imshow(content.T[:,20000:25000],aspect='auto',cmap=colorT)
plt.xticks([])
plt.yticks([])

handleT = [matplotlib.patches.Patch(color=colorTList[i], label=labelT[i]) for i in range(8)]
plt.legend(handles=handleT,loc='lower left',bbox_to_anchor=(1,0))

plt.tight_layout(pad=0.01)

plt.savefig("dsspTotalEW.png",dpi=300)

plt.show()
```
