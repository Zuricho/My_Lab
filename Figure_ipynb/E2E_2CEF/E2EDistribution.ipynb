```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
E2EEW_raw = np.loadtxt("distance_2cef_EW.dat",skiprows=1)
E2EEW = np.append(E2EEW_raw[1500:5000,1],[E2EEW_raw[6500:10000,1],E2EEW_raw[11500:15000,1],E2EEW_raw[16500:20000,1],E2EEW_raw[21500:25000,1]])

E2EEWCMAP_raw = np.loadtxt("distance_2cef_EW_CMAP.dat",skiprows=1)
E2EEWCMAP = np.append(E2EEWCMAP_raw[1500:5000,1],[E2EEWCMAP_raw[6500:10000,1],E2EEWCMAP_raw[11500:15000,1],E2EEWCMAP_raw[16500:20000,1],E2EEWCMAP_raw[21500:25000,1]])

E2ED_raw = np.loadtxt("distance_2cef_D.dat",skiprows=1)
E2ED = np.append(E2ED_raw[1500:5000,1],[E2ED_raw[6500:10000,1],E2ED_raw[11500:15000,1],E2ED_raw[16500:20000,1],E2ED_raw[21500:25000,1]])

E2EDCMAP_raw = np.loadtxt("distance_2cef_D_CMAP.dat",skiprows=1)
E2EDCMAP = np.append(E2EDCMAP_raw[1500:5000,1],[E2EDCMAP_raw[6500:10000,1],E2EDCMAP_raw[11500:15000,1],E2EDCMAP_raw[16500:20000,1],E2EDCMAP_raw[21500:25000,1]])

```

```python
plt.subplots(figsize=(6,3))

plt.title("End to End Distance 2CEF")

sns.kdeplot(E2EEW,shade=True,color='tab:blue',label='ffPTM TIP4PEW')
plt.axvline(x=sum(E2EEW)/len(E2EEW), ymin=0, ymax=1,color="tab:blue",linestyle="--")

sns.kdeplot(E2EEWCMAP,shade=True,color='tab:orange',label='ff03CMAP TIP4PEW')
plt.axvline(x=sum(E2EEWCMAP)/len(E2EEWCMAP), ymin=0, ymax=1,color="tab:orange",linestyle="--")

sns.kdeplot(E2ED,shade=True,color='tab:green',label='ffPTM TIP4PD')
plt.axvline(x=sum(E2ED)/len(E2ED), ymin=0, ymax=1,color="tab:green",linestyle="--")

sns.kdeplot(E2EDCMAP,shade=True,color='tab:red',label='ff03CMAP TIP4PD')
plt.axvline(x=sum(E2EDCMAP)/len(E2EDCMAP), ymin=0, ymax=1,color="tab:red",linestyle="--")

plt.legend()
plt.savefig("E2E_2cef.png",dpi=300)
plt.show()
```
