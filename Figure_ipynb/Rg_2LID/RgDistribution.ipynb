```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
RgSim1 = np.loadtxt("Rg_2lid_sim1.dat",skiprows=1)
RgSim2 = np.loadtxt("Rg_2lid_sim2.dat",skiprows=1)
RgSim3 = np.loadtxt("Rg_2lid_sim3.dat",skiprows=1)
RgSim4 = np.loadtxt("Rg_2lid_sim4.dat",skiprows=1)
RgSim5 = np.loadtxt("Rg_2lid_sim5.dat",skiprows=1)
Rg = np.append(RgSim1[1500:,1],[RgSim2[1500:,1],RgSim3[1500:,1],RgSim4[1500:,1],RgSim5[1500:,1]])


RgCMAPSim1 = np.loadtxt("Rg_2lid_CMAP_sim1.dat",skiprows=1)
RgCMAPSim2 = np.loadtxt("Rg_2lid_CMAP_sim2.dat",skiprows=1)
RgCMAPSim3 = np.loadtxt("Rg_2lid_CMAP_sim3.dat",skiprows=1)
RgCMAPSim4 = np.loadtxt("Rg_2lid_CMAP_sim4.dat",skiprows=1)
RgCMAPSim5 = np.loadtxt("Rg_2lid_CMAP_sim5.dat",skiprows=1)
RgCMAP = np.append(RgCMAPSim1[1500:,1],[RgCMAPSim2[1500:,1],RgCMAPSim3[1500:,1],RgCMAPSim4[1500:,1],RgCMAPSim5[1500:,1]])


RgDSim1 = np.loadtxt("Rg_2lid_D_sim1.dat",skiprows=1)
RgDSim2 = np.loadtxt("Rg_2lid_D_sim2.dat",skiprows=1)
RgDSim3 = np.loadtxt("Rg_2lid_D_sim3.dat",skiprows=1)
RgDSim4 = np.loadtxt("Rg_2lid_D_sim4.dat",skiprows=1)
RgDSim5 = np.loadtxt("Rg_2lid_D_sim5.dat",skiprows=1)
RgD = np.append(RgDSim1[1500:,1],[RgDSim2[1500:,1],RgDSim3[1500:,1],RgDSim4[1500:,1],RgDSim5[1500:,1]])


RgDCMAPSim1 = np.loadtxt("Rg_2lid_D_CMAP_sim1.dat",skiprows=1)
RgDCMAPSim2 = np.loadtxt("Rg_2lid_D_CMAP_sim2.dat",skiprows=1)
RgDCMAPSim3 = np.loadtxt("Rg_2lid_D_CMAP_sim3.dat",skiprows=1)
RgDCMAPSim4 = np.loadtxt("Rg_2lid_D_CMAP_sim4.dat",skiprows=1)
RgDCMAPSim5 = np.loadtxt("Rg_2lid_D_CMAP_sim5.dat",skiprows=1)
RgDCMAP = np.append(RgDCMAPSim1[1500:,1],[RgDCMAPSim2[1500:,1],RgDCMAPSim3[1500:,1],RgDCMAPSim4[1500:,1],RgDCMAPSim5[1500:,1]])
```

```python
plt.subplots(figsize=(6,3))

plt.title("Radius of Gyration 2LID")

sns.kdeplot(Rg,shade=True,color='tab:blue',label='ffPTM TIP4PEW')
plt.axvline(x=sum(Rg)/len(Rg), ymin=0, ymax=1,color="tab:blue",linestyle="--")

sns.kdeplot(RgCMAP,shade=True,color='tab:orange',label='ff03CMAP TIP4PEW')
plt.axvline(x=sum(RgCMAP)/len(RgCMAP), ymin=0, ymax=1,color="tab:orange",linestyle="--")

sns.kdeplot(RgD,shade=True,color='tab:green',label='ffPTM TIP4PD')
plt.axvline(x=sum(RgD)/len(RgD), ymin=0, ymax=1,color="tab:green",linestyle="--")

sns.kdeplot(RgDCMAP,shade=True,color='tab:red',label='ff03CMAP TIP4PD')
plt.axvline(x=sum(RgDCMAP)/len(RgDCMAP), ymin=0, ymax=1,color="tab:red",linestyle="--")

plt.legend()
plt.savefig("Rg_2lid.png",dpi=300)
plt.show()
```
