```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

def loadfile(fileName):
    content = np.loadtxt(fileName,skiprows=1)
    return content
    
```

```python
dihAngle_SEP = loadfile("phipsi_SEP.dat")
dihAngle_SEP_CMAP = loadfile("phipsi_SEP_CMAP.dat")

dihAngle_TPO = loadfile("phipsi_TPO.dat")
dihAngle_TPO_CMAP = loadfile("phipsi_TPO_CMAP.dat")

dihAngle_PTR = loadfile("phipsi_PTR.dat")
dihAngle_PTR_CMAP = loadfile("phipsi_PTR_CMAP.dat")
```

```python
plt.subplots(figsize=(15,10))

plt.subplot(231)
sns.kdeplot(dihAngle_SEP[:,1],dihAngle_SEP[:,2],shade=False,n_levels=10,cut=0,cmap="gnuplot2_r")
plt.xlim(-180,180)
plt.ylim(-180,180)

plt.subplot(234)
sns.kdeplot(dihAngle_SEP_CMAP[:,1],dihAngle_SEP_CMAP[:,2],shade=False,n_levels=10,cut=0,cmap="gnuplot2_r")
plt.xlim(-180,180)
plt.ylim(-180,180)

plt.subplot(232)
sns.kdeplot(dihAngle_TPO[:,1],dihAngle_TPO[:,2],shade=False,n_levels=10,cut=0,cmap="gnuplot2_r")
plt.xlim(-180,180)
plt.ylim(-180,180)

plt.subplot(235)
sns.kdeplot(dihAngle_TPO_CMAP[:,1],dihAngle_TPO_CMAP[:,2],shade=False,n_levels=10,cut=0,cmap="gnuplot2_r")
plt.xlim(-180,180)
plt.ylim(-180,180)

plt.subplot(233)
sns.kdeplot(dihAngle_PTR[:,1],dihAngle_PTR[:,2],shade=False,n_levels=10,cut=0,cmap="gnuplot2_r")
plt.xlim(-180,180)
plt.ylim(-180,180)

plt.subplot(236)
sns.kdeplot(dihAngle_PTR_CMAP[:,1],dihAngle_PTR_CMAP[:,2],shade=False,n_levels=10,cut=0,cmap="gnuplot2_r")
plt.xlim(-180,180)
plt.ylim(-180,180)



plt.savefig("Ramaplot.png",dpi=500)
plt.show()
```
